## Question 3
### (a) Using the pandas library, read the csv file and create a data-frame called ufc.

```python
# Importing libraries
import pandas as pd

# Reading csv file
ufc = pd.read_csv("UFC_data.csv")
ufc.head()
```

### (b) Using the appropriate commands, remove all the observations in which weight class is equal to OpenWeight.

```python
ufc = ufc[ufc['weight_class'] != 'OpenWeight']
ufc.head()
```

### (c) Using the appropriate commands, report the corner with highest winning rate (is it Red, Blue or draw?)

```python
ufc['Winner'].value_counts() / ufc.shape[0]
```

##### The corner with highest winning rate is Red.

### (d) Using the appropriate commands, report the corner with highest winning rate (is it Red, Blue or draw?) across the different weight classes. Is this consistent with your answer from part (c)? Explain.

```python
ufc.groupby('weight_class')['Winner'].value_counts() / ufc.groupby('weight_class')['Winner'].count()
```

#### Yes, my answer from part (c) still consistent because the corner with highest winning rate is Red, even breaking it down on weight classes.

### (e) Using the appropriate commands, report the corner with highest winning rate (is it Red, Blue or draw?) across the different weight classes and whether or not the fight is a title bout. Is this consistent with your answers from parts (c)-(d)? Explain.

```python
ufc.groupby(['weight_class', 'title_bout'])['Winner'].value_counts() / ufc.groupby(['weight_class', 'title_bout'])['Winner'].count()
```

##### I reamain with the same answer from the previous questions. Red corners still with the highest winning rate.

### (f) Using your answers from parts (c)-(e), what can you conclude from that analysis? Explain.
